```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x[0]**2-2*x[0]+x[1]   
```

```python
x = np.linspace(0, 3, 100)
y = np.linspace(0, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f(np.vstack([X.ravel(), Y.ravel()])).reshape((100,100))
plt.figure(figsize=(10,8))
plt.contour(X, Y, Z, np.arange(-10,10, 0.5), cmap='jet');
plt.plot(x, 1-x, 'k:', linewidth=1)
plt.plot(x, (2-x)/4, 'k:', linewidth=1)
plt.xlabel(''); plt.ylabel(''); 
plt.title("Proyecci√≥n en el plano XY")
plt.axis([0,3,0,3])
```

```python
cons = ({'type': 'ineq',
         'fun' : lambda x: np.array([-x[0]-4*x[1]+2],([x[0]+x[1]-1]))})

bnds = ((0, 2.0), (0, 2.0))
```

```python

x0 = [0.5, 0.5]
```

```python

cx = opt.minimize(f, x0, bounds=bnds, constraints=None)
cx
```
