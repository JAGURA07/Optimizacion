Maximo y minimo de la funcion $f(x)=xy+x+y$ usando diferentes metodos

```python
import numpy as np
from scipy.optimize import minimize
```

*Maximo*

```python
def funcion(x):
    return ((x[0]*x[1])+x[0]+x[1])*1
```

```python
x0 = np.array([1.5, 2.5])
res = minimize(funcion, x0, method='nelder-mead',
               bounds=([1, 2],[2, 3]),
               options={'xatol': 1e-8, 'disp': True})
```

```python
print(res.x)
```

*Minimo*

```python
def funcion(x):
    return ((x[0]*x[1])+x[0]+x[1])*-1
```

```python
x0 = np.array([1.5, 2.5])
res = minimize(funcion, x0, method='nelder-mead',
               bounds=([1, 2],[2, 3]),
               options={'xatol': 1e-8, 'disp': True})
```

```python
print(res.x)
```
