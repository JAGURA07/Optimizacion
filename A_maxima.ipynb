```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return x/2-x*x
```

```python
x = np.arange(0,1,0.01)
```

```python
plt.plot(x, [f(i) for i in x])
plt.xlim(0, 0.5)
plt.ylim(0, 0.07)
plt.title("Area Maxima")
plt.text(0.25,0.0625,"")
plt.plot(0.25,0.0625, 'r--o', label="Part√≠cula 1")
plt.show()
```
