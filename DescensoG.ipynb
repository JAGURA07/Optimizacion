```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
x = np.linspace(-4, 4, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
def z(x,y):
  return 100*(y-x**2)**2+(1-x)**2
ax.plot_surface(X, Y, z(X,Y), cmap='viridis', alpha=0.6)
plt.show()
```

```python
def f(x):
  return 100*(x[1]-x[0]**2)**2+(1-x[0])**2

def Gf(x):
  G=np.zeros(2)
  G[0]=-400*x[0]*(x[1]-x[0]**2)-2*(1-x[0])
  G[1]=200*(x[1]-x[0]**2)
  return G

def descenso(f,Gf,x0,t,n):
  x=x0
  iteraciones=[]
  y=[]
  for i in range(n):
    x=x-t*Gf(x)
    y.append(f(x))
    iteraciones.append(i)
    print('x = ', x, ', y = ', f(x))
  plt.plot(iteraciones,y)

descenso(f,Gf,[2,1],0.0001,70)
```
